### Sending a GET request

```python
# Define the base URL
# Base URL: the part of the URL common to all requests, not containing the parameters
base_url = "https://rest.coinapi.io/v1/exchangerate/BTC?apikey=830D90AE-3D06-4CCA-B63F-BAE4C69AB0DC"
#830D90AE-3D06-4CCA-B63F-BAE4C69AB0DC
```

```python
# We can make a GET request to this API endpoint with requests.get
url_btc_usd_2020 = 'https://rest.coinapi.io/v1/exchangerate/BTC/USD/history?period_id=10DAY&time_start=2020-01-01T00:00:00&time_end=2021-01-01T00:00:00'
headers = {'X-CoinAPI-Key' : '830D90AE-3D06-4CCA-B63F-BAE4C69AB0DC'}
response = requests.get(url_btc_usd_2020, headers=headers)
```

### Investigating the response

```python
# Checking the status code of the response
response.status_code  # 200 means everything is ok  somethine like 404, those are status codes
```

```python
# Inspecting the content body of the response as text or string
response.text
```

```python
# Inspecting the content of the response in bytes format it will send
response.content
```

### Incorporating parameters in the GET request

```python
# Making a request to the server with the new URL, containing the parameters
headers = {'X-CoinAPI-Key' : '830D90AE-3D06-4CCA-B63F-BAE4C69AB0DC'}
response2 = requests.get(param_url, headers=headers)
response2.status_code
```
