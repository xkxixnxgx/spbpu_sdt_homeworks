```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from matplotlib.ticker import PercentFormatter
```

### BARCHART

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
plt.figure(figsize = (20, 10))


mlist=["black","white","pink","purple","red","green","blue"]
plt.barh(df_used_cars["Brand"],
        df_used_cars["Cars Listings"],
        color = mlist)

plt.xticks(rotation = 45,fontsize = 20, color = 'green')
plt.yticks(rotation = 45, fontsize = 20, color = 'red')
plt.title("Cars Listings by Viktor", fontsize = 40, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 20 )
plt.show()
```

### HISTOGRAM

```python
df_real_estate = pd.read_csv("histogram_survey_data.csv")
df_real_estate.head(5)
```

```python
sns.set_style("white") 
plt.figure(figsize = (10, 6))
plt.hist(df_real_estate["Age"],
         bins = 6,
         color = "#CC6600")
plt.title("Age groups of the participants in the survey", fontsize = 20, weight = "bold")
plt.xlabel("Age",fontsize=15)  
plt.ylabel("Number of People",fontsize=15)
sns.despine()

plt.show()
```

### SCATTER PLOT

```python
df_real_estate = pd.read_csv("student_scores_data.csv")
df_real_estate.head(5)
```

```python
plt.figure(figsize = (12, 4))
sns.scatterplot(
    x = df_real_estate['SAT'],
    y = df_real_estate['GPA'],
    s = 200,
)
plt.title("GPA and SAT student scores", 
          fontsize = 15,
          weight = "bold")
plt.xlabel("SAT", fontsize=15)
plt.ylabel("GPA", fontsize=15)
plt.show()
```

### LINECHART

```python
df_returns = pd.read_csv("returns1020.csv")
df_returns.head(5)
```

```python
df_returns["new_date"] = pd.to_datetime(df_returns["Date"])
df_returns["new_date"].head(5)
```

```python
df_returns_from_2010_to_2020 = df_returns[(df_returns.new_date >= '2010-01-01') &
                                      (df_returns.new_date <= '2020-12-31')]
```

```python
df_returns_from_2010_to_2020.head(5)
```

```python
plt.figure(figsize = (10, 8))

labels = ["GSPCRet", "FTSERet"]
plt.plot(df_returns_from_2010_to_2020["new_date"], df_returns_from_2010_to_2020["GSPCRet"], color = "purple")
plt.plot(df_returns_from_2010_to_2020["new_date"], df_returns_from_2010_to_2020["FTSERet"], color = "green")
plt.title("GSPCRet vs FTSERet (2010-2020)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns",fontsize = 20)
plt.xlabel("Years",fontsize = 20)
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
df_returns_from_2018_to_2020 = df_returns[(df_returns.new_date >= '2018-01-01') &
                                      (df_returns.new_date <= '2020-12-31')]
```

```python
plt.figure(figsize = (10, 8))

labels = ["GSPCRet", "FTSERet"]
plt.plot(df_returns_from_2018_to_2020["new_date"], df_returns_from_2018_to_2020["GSPCRet"], color = "purple")
plt.plot(df_returns_from_2018_to_2020["new_date"], df_returns_from_2018_to_2020["FTSERet"], color = "green")
plt.title("GSPCRet vs FTSERet (2018-2020)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns",fontsize = 20)
plt.xlabel("Years",fontsize = 20)
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

### STACKED AREA CHART

```python
df_returns = pd.read_csv("area_chart_homework_data.csv")
df_returns.head(12)
```

```python
df_returns['month'] = pd.DatetimeIndex(df_returns['Date']).month_name()
df_returns.head(12)
```

```python
label_x = df_returns['month'].to_list()
plt.figure(figsize = (12, 6))

plt.stackplot(
    df_returns["month"],
    df_returns["Domestic"],
    df_returns["International"],
    colors =['r', 'c']
)
plt.xticks(df_returns["month"], rotation = 45)
plt.legend(labels = ["Domestic", "International"], loc='upper left')
plt.xlabel("Months", fontsize = 15)
plt.ylabel("Number of Sales", fontsize = 15)
plt.title("Number of Sales in 2020", fontsize = 20, weight = "bold")
sns.despine()
plt.show()
```
